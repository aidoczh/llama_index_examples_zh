```python
### DatabaseReader ç±» ###
# db æ˜¯ DatabaseReader çš„ä¸€ä¸ªå®ä¾‹ï¼š
print(type(db))
# DatabaseReader å¯ç”¨çš„æ–¹æ³•ï¼š
print(type(db.load_data))

### SQLDatabase ç±» ###
# db.sql æ˜¯ SQLDatabase çš„ä¸€ä¸ªå®ä¾‹ï¼š
print(type(db.sql_database))
# SQLDatabase å¯ç”¨çš„æ–¹æ³•ï¼š
print(type(db.sql_database.from_uri))
print(type(db.sql_database.get_single_table_info))
print(type(db.sql_database.get_table_columns))
print(type(db.sql_database.get_usable_table_names))
print(type(db.sql_database.insert_into_table))
print(type(db.sql_database.run_sql))
# SQLDatabase å¯ç”¨çš„å±æ€§ï¼š
print(type(db.sql_database.dialect))
print(type(db.sql_database.engine))
```

```python
### æµ‹è¯• DatabaseReader
### æ¥è‡ª SQLDatabaseï¼ŒSQLAlchemy å¼•æ“å’Œæ•°æ®åº“ URIï¼š

# ä» SQLDatabase å®ä¾‹ä¸­ï¼š
print(type(db.sql_database))
db_from_sql_database = DatabaseReader(sql_database=db.sql_database)
print(type(db_from_sql_database))

# ä» SQLAlchemy å¼•æ“ä¸­ï¼š
print(type(db.sql_database.engine))
db_from_engine = DatabaseReader(engine=db.sql_database.engine)
print(type(db_from_engine))

# ä»æ•°æ®åº“ URI ä¸­ï¼š
print(type(db.uri))
db_from_uri = DatabaseReader(uri=db.uri)
print(type(db_from_uri))
```

```python
# ä»¥ä¸‹SQLæŸ¥è¯¢ç¤ºä¾‹è¿”å›ç”¨æˆ·è¡¨ä¸­å¹´é¾„å¤§äºæˆ–ç­‰äº18å²çš„æ¯è¡Œçš„åç§°å’Œå¹´é¾„åˆ—çš„è¿æ¥æ–‡æœ¬å€¼åˆ—è¡¨

query = f"""
    SELECT
        CONCAT(name, ' is ', age, ' years old.') AS text
    FROM public.users
    WHERE age >= 18
    """
```

```python
# è¯·å‚è€ƒllama_index.utilities.sql_wrapper
# SQLDatabase.run_sqlæ–¹æ³•
texts = db.sql_database.run_sql(command=query)

# æ˜¾ç¤ºtype(texts)å’Œtexts
# type(texts)å¿…é¡»è¿”å›<class 'list'>
print(type(texts))

# æ–‡æ¡£å¿…é¡»è¿”å›ä¸€ä¸ªå…ƒç»„å¯¹è±¡çš„åˆ—è¡¨
print(texts)
```

```python
# è¯·å‚è€ƒllama_index.readers.database.DatabaseReader.load_data
# DatabaseReader.load_dataæ–¹æ³•
documents = db.load_data(query=query)

# æ˜¾ç¤ºtype(documents)å’Œdocuments
# type(documents)å¿…é¡»è¿”å›<class 'list'>
print(type(documents))

# documentså¿…é¡»è¿”å›ä¸€ä¸ªDocumentå¯¹è±¡çš„åˆ—è¡¨
print(documents)
```
