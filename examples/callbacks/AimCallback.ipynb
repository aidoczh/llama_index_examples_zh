```python
from llama_index.core.callbacks import CallbackManager
from llama_index.callbacks.aim import AimCallback
from llama_index.core import SummaryIndex
from llama_index.core import SimpleDirectoryReader
```

è®©æˆ‘ä»¬ä½¿ç”¨`SimpleDirectoryReader`ä»'examples/data/paul_graham'ä¸­è¯»å–æ–‡æ¡£ã€‚


#### ä¸‹è½½æ•°æ®


```python
!mkdir -p 'data/paul_graham/'
!wget 'https://raw.githubusercontent.com/run-llama/llama_index/main/docs/docs/examples/data/paul_graham/paul_graham_essay.txt' -O 'data/paul_graham/paul_graham_essay.txt'
```

```python
docs = SimpleDirectoryReader("./data/paul_graham").load_data()
```

ç°åœ¨è®©æˆ‘ä»¬åˆå§‹åŒ–ä¸€ä¸ªAimCallbackå®ä¾‹ï¼Œå¹¶å°†å…¶æ·»åŠ åˆ°å›è°ƒç®¡ç†å™¨çš„åˆ—è¡¨ä¸­ã€‚


```python
aim_callback = AimCallback(repo="./")
callback_manager = CallbackManager([aim_callback])
```

åœ¨è¿™ä¸ªç‰‡æ®µä¸­ï¼Œæˆ‘ä»¬åˆå§‹åŒ–äº†ä¸€ä¸ªå›è°ƒç®¡ç†å™¨ã€‚
æ¥ä¸‹æ¥ï¼Œæˆ‘ä»¬é€šè¿‡ä¼ å…¥æ–‡æ¡£é˜…è¯»å™¨å’Œå›è°ƒå‡½æ•°æ¥åˆ›å»º`SummaryIndex`ç±»çš„ä¸€ä¸ªå®ä¾‹ã€‚ç„¶åæˆ‘ä»¬åˆ›å»ºä¸€ä¸ªæŸ¥è¯¢å¼•æ“ï¼Œç”¨äºåœ¨ç´¢å¼•ä¸Šè¿è¡ŒæŸ¥è¯¢å¹¶æ£€ç´¢ç›¸å…³ç»“æœã€‚


```python
index = SummaryIndex.from_documents(docs, callback_manager=callback_manager)
query_engine = index.as_query_engine()
```

æœ€åè®©æˆ‘ä»¬æ ¹æ®æˆ‘ä»¬æä¾›çš„æ–‡æ¡£å‘è¯­è¨€æ¨¡å‹æå‡ºä¸€ä¸ªé—®é¢˜ã€‚


```python
response = query_engine.query("What did the author do growing up?")
```
