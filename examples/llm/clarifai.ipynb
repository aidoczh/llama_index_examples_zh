```python
# 示例参数params = dict(    user_id="clarifai",    app_id="ml",    model_name="llama2-7b-alternative-4k",    model_url=(        "https://clarifai.com/clarifai/ml/models/llama2-7b-alternative-4k"    ),)
```

```python
# 方法：1 使用 model_url 参数llm_model = Clarifai(model_url=params["model_url"])
```

```python
# 方法2：使用model_name、app_id和user_id参数llm_model = Clarifai(    model_name=params["model_name"],    app_id=params["app_id"],    user_id=params["user_id"],)
```

```python
llm_reponse = llm_model.complete(
    prompt="write a 10 line rhyming poem about science"
)
```

调用`chat`函数


```python
from llama_index.core.llms import ChatMessage

messages = [
    ChatMessage(role="user", content="write about climate change in 50 lines")
]
Response = llm_model.chat(messages)
```

```python
print(Response)
```
