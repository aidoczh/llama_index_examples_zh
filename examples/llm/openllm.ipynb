```python
import os
from typing import List, Optional

from llama_index.llms.openllm import OpenLLM, OpenLLMAPI
from llama_index.core.llms import ChatMessage
```

```python
# è¿™é‡Œä½¿ç”¨äº† https://huggingface.co/HuggingFaceH4/zephyr-7b-alpha# å¦‚æœæ˜¯ç¬¬ä¸€æ¬¡è°ƒç”¨ï¼Œåˆ™ä¼šä¸‹è½½åˆ°æœ¬åœ° Hugging Face æ¨¡å‹ç¼“å­˜ä¸­ï¼Œ# ç„¶ååœ¨æœ¬åœ°æœºå™¨ä¸Šå®é™…è¿è¡Œæ¨¡å‹local_llm = OpenLLM("HuggingFaceH4/zephyr-7b-alpha")# è¿™å°†ä½¿ç”¨è¿è¡Œåœ¨ localhost:3000 æœåŠ¡å™¨ä¸Šçš„æ¨¡å‹remote_llm = OpenLLMAPI(address="http://localhost:3000")# æ³¨æ„è¿™é‡Œå¦‚æœè®¾ç½®äº† OPENLLM_ENDPOINT ç¯å¢ƒå˜é‡ï¼Œåˆ™ä¸éœ€è¦ä¼ å…¥åœ°å€# å¦‚æœæ²¡æœ‰ä¼ å…¥åœ°å€ï¼Œåˆ™åœ°å€ä¸º address=os.getenv("OPENLLM_ENDPOINT")remote_llm = OpenLLMAPI()
```

ä½¿ç”¨`OpenLLM`ä½œä¸ºå®Œæˆçš„åŸºç¡€æ”¯æŒä¸[vLLM](https://vllm.ai/)ä¸€èµ·è¿›è¡Œè¿ç»­çš„æ‰¹å¤„ç†ã€‚


```python
completion_response = remote_llm.complete("To infinity, and")
print(completion_response)
```

`OpenLLM`å’Œ`OpenLLMAPI`è¿˜æ”¯æŒ`complete`çš„æµå¼ä¼ è¾“ã€åŒæ­¥å’Œå¼‚æ­¥æ“ä½œï¼š


```python
for it in remote_llm.stream_complete(
    "The meaning of time is", max_new_tokens=128
):
    print(it, end="", flush=True)
```

```python
async for it in remote_llm.astream_chat(
    [
        ChatMessage(
            role="system", content="You are acting as Ernest Hemmingway."
        ),
        ChatMessage(role="user", content="Hi there!"),
        ChatMessage(role="assistant", content="Yes?"),
        ChatMessage(role="user", content="What is the meaning of life?"),
    ]
):
    print(it.message.content, flush=True, end="")
```
