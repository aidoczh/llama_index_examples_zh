# WatsonX


## 基本用法


#### 使用提示调用`complete`


```python
%pip install llama-index-llms-watsonx
```

```python
from llama_index.llms.watsonx import WatsonX

credentials = {
    "url": "https://enter.your-ibm.url",
    "apikey": "insert_your_api_key",
}

project_id = "insert_your_project_id"

resp = WatsonX(credentials=credentials, project_id=project_id).complete(
    "Paul Graham is"
)
```

```python
print(resp)
```

#### 使用消息列表调用`chat`


```python
from llama_index.core.llms import ChatMessage
from llama_index.llms.watsonx import WatsonX


messages = [
    ChatMessage(
        role="system", content="You are a pirate with a colorful personality"
    ),
    ChatMessage(role="user", content="Tell me a story"),
]

resp = WatsonX(
    model_id="meta-llama/llama-2-70b-chat",
    credentials=credentials,
    project_id=project_id,
).chat(messages)
```

```python
print(resp)
```

## 流式处理


使用`stream_complete`端点


```python
from llama_index.llms.watsonx import WatsonX

llm = WatsonX(credentials=credentials, project_id=project_id)

resp = llm.stream_complete("Paul Graham is")
```

```python
for r in resp:
    print(r.delta, end="")
```

使用 `stream_chat` 终端点


```python
from llama_index.llms.watsonx import WatsonX

llm = WatsonX(
    model_id="meta-llama/llama-2-70b-chat",
    credentials=credentials,
    project_id=project_id,
)
messages = [
    ChatMessage(
        role="system", content="You are a pirate with a colorful personality"
    ),
    ChatMessage(role="user", content="Tell me a story"),
]

resp = llm.stream_chat(messages)
```

```python
for r in resp:
    print(r.delta, end="")
```

## 配置模型


```python
from llama_index.llms.watsonx import WatsonX

llm = WatsonX(
    model_id="meta-llama/llama-2-70b-chat",
    credentials=credentials,
    project_id=project_id,
    temperature=0,
    max_new_tokens=100,
)
```

```python
resp = llm.complete("Paul Graham is")
```

```python
print(resp)
```
