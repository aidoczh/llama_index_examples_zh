```python
from llama_index.core import SimpleDirectoryReader

# 加载一些示例数据
reader = SimpleDirectoryReader(input_files=["paul_graham_essay.txt"])
documents = reader.load_data()
```

### 启动 Ollama

确保你在终端中运行 `ollama run llama2`。


```python
# !ollama运行llama2
```

### 下载和初始化包

我们使用`download_llama_pack`来下载包类，然后我们用文档来初始化它。

每个包将有不同的初始化参数。您可以通过其[README](https://github.com/logan-markewich/llama-hub/tree/main/llama_hub/llama_packs/voyage_query_engine)（也在LlamaHub上）找到有关每个包的初始化参数的更多信息。

**注意**：您还必须指定一个输出目录。在这种情况下，包被下载到`voyage_pack`。这允许您自定义和对文件进行更改，并稍后导入它！


```python
from llama_index.core.llama_pack import download_llama_pack

# 下载并安装依赖项
OllamaQueryEnginePack = download_llama_pack(
    "OllamaQueryEnginePack", "./ollama_pack"
)
```

```python
# 您可以使用任何llama-hub加载器来获取文档！
ollama_pack = OllamaQueryEnginePack(model="llama2", documents=documents)
```

```python
response = ollama_pack.run("What did the author do growing up?")
```
