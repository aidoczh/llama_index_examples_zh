## 使用数据加载器

在这个例子中，我们将展示如何使用`SimpleWebPageReader`。

**注意**：对于LlamaHub上的任何模块，如果要与`download_`函数一起使用，请记下类名。


```python
%pip install llama-index-agent-openai
%pip install llama-index-readers-web
%pip install llama-index-tools-google
```

```python
from llama_index.readers.web import SimpleWebPageReader
```

```python
reader = SimpleWebPageReader(html_to_text=True)
```

```python
docs = reader.load_data(urls=["https://eugeneyan.com/writing/llm-patterns/"])
```

```python
print(docs[0].get_content()[:400])
```

现在你可以将这些文档插入到你的下游LlamaIndex流水线中。


```python
from llama_index.core import VectorStoreIndex

index = VectorStoreIndex.from_documents(docs)
query_engine = index.as_query_engine()
```

```python
response = query_engine.query("What are ways to evaluate LLMs?")
print(str(response))
```

## 使用代理工具规范

在这个例子中，我们将展示如何加载一个代理工具。


```python
from llama_index.tools.google import GmailToolSpec
```

```python
tool_spec = GmailToolSpec()
```

```python
# 连接到你的代理from llama_index.agent.openai import OpenAIAgent
```

```python
agent = OpenAIAgent.from_tools(tool_spec.to_tool_list())
```

```python
agent.chat("What is my most recent email")
```
